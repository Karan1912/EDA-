```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
x = data.drop('PROFIT',axis=1)
```

```python
c = encoder.fit_transform(x.STATE)
```

```python
c
```

```python
x["STATE"] = c 
```

```python
x.head()
```

```python
x.values
```

```python
y = data['PROFIT'].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_pred = reg.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_pred,y_test)
```

```python
mse
```

```python
print("weight:",reg.coef_)
print("Intercept:",reg.intercept_)
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test,y_pred)
```

```python
r2_score
```

```python
 reg_test =reg.score(x_test, y_test)
 print('Model performance on Test data: ',reg_test)
```

```python
reg_train = reg.score(x_train, y_train)
print('Model performance on Train data: ',reg_train)
```

```python
residuals = y_test-y_pred
mean_residuals = np.mean(residuals)
print("Mean of Residuals {}".format(mean_residuals))
```
